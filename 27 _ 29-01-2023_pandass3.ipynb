```python
d1 = {"key1" : [1,2,3,4,5,"sudh"],
      "key2" : [7,8,9,10,11,12],
      "key3" : [13,14,15,16,17,18],
      "key4" : [19,20,21,22,23,24]        
}
```

```python
import pandas as pd
```

Coverting Dictionary to data frame

```python
pd.DataFrame(d1) #if dictionary d1 have not similar length of data thet it will throw error
```

initialize our own index also

```python
df1 = pd.DataFrame(d1, index=["a","b","c", "d","e","f"]) #if dictionary d1 have not similar length of data thet it will throw error
```

```python
d2 = {"key1" : [1,2,3,4,5,"sudh"],
      "key2" : [7,8,"kumar",10,11,12],
      "key3" : [13,14,15,16,17,18],
      "key4" : [19,20,21,"iNeuron","pwskills",24]        
}
```

```python
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

Merging two dataframe

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how="inner", on="key1")
#How="inner" : Data Which is available in both of the row at the same index, is merged and give output of that datas, by defaul it is inner
#on="*column_name" : which column we want to check
```

```python
pd.merge(df1,df2, how="inner", on="key4")
#How="inner" : Data Which is available in both of the row at the same index, is merged and give output of that datas, by defaul it is inner
#on="*column_name" : which column we want to check, means if which data is available in d1 key4 column and also available in d2 key4 column this is given
```

```python
pd.merge(df1,df2, how="left", on="key4")
```

```python
pd.merge(df1,df2, how="right", on="key4")
```

```python
pd.merge(df1,df2, how="outer", on="key4")
```

Merging based on two keys

```python
pd.merge(df1,df2, how="outer", on=["key4", "key1"])
```

Performing Merging Operation on two dataframe with different index name, if we have that situation we have to use right_on and left_on parameter

```python
d11 = {"a" : [1,2,3,4,5,"sudh"],
      "b" : [7,8,9,10,11,12],
      "c" : [13,14,15,16,17,18],
      "d" : [19,20,21,22,23,24]        
}
```

```python
d12 = {"key1" : [1,2,3,4,5,"sudh"],
      "key2" : [7,8,"kumar",10,11,12],
      "key3" : [13,14,15,16,17,18],
      "key4" : [19,20,21,"iNeuron","pwskills",24]        
}
```

```python
df11 = pd.DataFrame(d11)
```

```python
df11
```

```python
df12 = pd.DataFrame(d12)
```

```python
df12
```

```python
pd.merge(df11, df12, how="left", left_on= "a", right_on="key1" )
```
